```python
tasks = ["email", "meeting", "write report", "lunch", "coding", "review"]
#           0        1           2          3        4         5

# Get tasks from index 1 up to (but not including) 4
morning_tasks = tasks[1:4]
print(morning_tasks)  # Output: ['meeting', 'write report', 'lunch']
```

```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(f"I like {fruit}s!")
# Output:
# I like apples!
# I like bananas!
# I like cherries!
```

```python
numbers = [1, 2, 3, 4, 5]
# Create a new list with the square of each number
squares = [n**2 for n in numbers]
print(squares)  # Output: [1, 4, 9, 16, 25]
```

```python
numbers = [1, 2, 3, 4, 5, 6]
# Create a new list with the squares of only the even numbers
even_squares = [n**2 for n in numbers if n % 2 == 0]
print(even_squares)  # Output: [4, 16, 36]
```
